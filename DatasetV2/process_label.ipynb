```python
import pandas as pd
```

```python
df = pd.read_csv('old_label.csv')
df.head()
```

```python
# Melt the DataFrame to the desired format
df_melted = df.melt(id_vars='tps', var_name='paslon', value_name='suara')
df_melted.head()
```

```python
df_melted['tps'] = df_melted['tps'] + "_" + df_melted['paslon'].astype(str)
df_melted
```

```python
df_melted.drop(columns=['paslon'], inplace=True)
```

```python
df_melted.sort_values(by=['tps'],inplace=True)
```

```python
df_melted
```

```python
df_melted.to_csv('label.csv', index=False)
```
