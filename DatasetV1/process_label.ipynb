```python
import pandas as pd
```

```python
label_df = pd.read_csv('label.csv')
label_df
```

```python
label_df.melt(id_vars='tps', var_name='paslon', value_name='suara')
```

```python
def add_new_rows(df):
    new_rows = []
    for index, row in df.iterrows():
        new_rows.append([row['tps'],row['suara']])
        new_rows.append([row['tps'] + '_blur',row['suara']])
        new_rows.append([row['tps'] + '_bright',row['suara']])
        new_rows.append([row['tps'] + '_contrass',row['suara']])
        new_rows.append([row['tps'] + '_glass',row['suara']])
        new_rows.append([row['tps'] + '_random',row['suara']])
    new_df = pd.DataFrame(new_rows, columns=df.columns)
    return new_df
```

```python
# Use the function to add new rows to the label_df
new_label_df = add_new_rows(label_df)
new_label_df
```

```python
new_label_df.to_csv('new_label.csv', index=False)
```
